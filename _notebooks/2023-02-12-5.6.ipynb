# Big Idea 5.5 Legal and Ethical Concerns 

### When creating a new github repository...
When someone creates a new git hub repository, they are asked to choose a license. Sme of the options are:
1. __MIT__. This is a permissive license. It allows others to use the code for any purpose, including commercial purposes, and to modify it. It also provides an express grant of patent rights from the contributor to the user. The MIT license is short and to the point. It lets people do anything they want with your code with proper attribution and without warranty.

2. __GPL__ is a copyleft license. It allows others to use the code for any purpose, including commercial purposes, and to modify it. It also provides an express grant of patent rights from the contributor to the user. The GPL license is short and to the point. It lets people do anything they want with your code with proper attribution and without warranty.

3. __Creative Commons licenses__ are a set of licenses that allow creators to choose how they want their work shared. There are six different Creative Commons licenses, each with different conditions. The most popular license is the Attribution-ShareAlike license, which allows others to distribute, remix, tweak, and build upon your work, even commercially, as long as they credit you for the original creation. This is the license used by Wikipedia, and is recommended for materials that would benefit from incorporating content from Wikipedia and similarly licensed projects.

### Good and bad passwords and what can help authentication

Good passwords:

- Long: A longer password is generally considered more secure than a short one. A password of at least 12 characters is recommended, but longer is even better.

- Complex: A complex password contains a mixture of uppercase and lowercase letters, numbers, and symbols. This makes it much more difficult for an attacker to guess or crack the password.

- Unique: A password should not be used for multiple accounts. If an attacker gains access to one password, they will have access to all of the other accounts as well.

Bad Passwords:

- Short: Short passwords are easy to guess or crack and should be avoided.

- Simple: Simple passwords, such as "password" or "123456", can be easily 
cracked and should be avoided.

- common: Common passwords, such as "qwerty" or "letmein", are well-known and easily cracked, and should be avoided.

Another step that can assist in authentication is the use of Two-Factor Authentication (2FA). With 2FA, in addition to entering a password, a user must also provide a second piece of information, such as a code sent to their phone or a biometric scan. This provides an extra layer of security and makes it much more difficult for an attacker to gain access to an account.

### symmetrical vs asymmetrical encryption

Symmetric encryption is a type of encryption where the same key is used to both encrypt and decrypt data. This means that if the key is lost, the data cannot be decrypted, so it is important to keep the key secure. An example of symmetric encryption is the Advanced Encryption Standard (AES).

Asymmetric encryption is a type of encryption where two different keys are used: a public key to encrypt data, and a private key to decrypt it. This means that the private key does not need to be shared, making it more secure. An example of asymmetric encryption is the RSA algorithm.

In AWS deployment, one example of encryption is the use of Amazon S3 server-side encryption. With this, data is automatically encrypted on the server before it is stored, and decrypted when it is retrieved. The encryption keys are managed by AWS, making it easier to use encryption in the cloud.

# Phishing and personal experience

A phishing scheme that I have learned about the hard way is one where attackers send an email that appears to be from a trusted source, such as a bank or a government agency. The email asks the recipient to click on a link and enter sensitive information, such as a password or credit card number. In reality, the link leads to a fake website that is controlled by the attacker, and the information entered is captured and used for malicious purposes. I remember being around 10 years old and then looking at an email saying that I won a $100 gift card. I got really happy and I entered click on the email and was led to a survey. Now, I am lucky that this wasn't a extremely harmful phishing scheme. This just collected a lot of data I entered and sold it to other companies. I don't remember if I entered any sensitive information, but I do remember that I wasn't very happy when my dad told me that it was a phishing scheme.

Other phishing techniques include:

- Social Engineering: This involves tricking people into revealing sensitive information through conversations or interactions. For example, an attacker may call a person and pretend to be a customer support representative, and ask for sensitive information.

- URL Spoofing: This involves creating a fake website that looks similar to a legitimate one, and tricking people into entering sensitive information.

- Malware: This involves installing malicious software on a person's computer that can steal sensitive information or control the machine.
